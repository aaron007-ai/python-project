```python
import random
import matplotlib.pyplot as plt
import pandas as pd

while True:
    print("\n=== Dice Roll Probability Simulator ===")
    num_rolls = int(input("Enter how many times to roll the dice: "))

    # Step 1: Simulate rolls
    rolls = [random.randint(1, 6) for _ in range(num_rolls)]

    # Step 2: Frequency count (1â€“6)
    frequency = {i: rolls.count(i) for i in range(1, 7)}

    # Step 3: Probability %
    probability = {i: (frequency[i] / num_rolls) * 100 for i in range(1, 7)}

    # Step 4: Display frequency
    print("\n--- Frequency Count ---")
    for i in range(1, 7):
        print(f"{i} appeared {frequency[i]} times")

    # Step 5: Display probability
    print("\n--- Probability Distribution ---")
    for i in range(1, 7):
        print(f"{i}: {probability[i]:.2f}%")

    # Step 6: Plot graph
    plt.bar(list(probability.keys()), list(probability.values()))
    plt.xlabel("Dice Number")
    plt.ylabel("Probability (%)")
    plt.title("Dice Roll Probability Distribution")
    plt.show()

    # Step 7: Export results
    df = pd.DataFrame({
        "Dice Number": list(frequency.keys()),
        "Frequency": list(frequency.values()),
        "Probability (%)": list(probability.values())
    })

    df.to_csv("results.csv", index=False)
    print("\nCSV file saved as results.csv!")

    # Step 8: Ask user to continue
    again = input("\nDo you want to run again? (y/n): ").lower()
    if again != "y":
        print("Exiting the program. Thank you!")
        break
```
